```python
### MATPLOTLIB - visualizations in Python
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [10,20,25,30]

plt.plot(x,y)
plt.show()
```

```python
plt.title("Simple Line Plot")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.plot(x,y)
plt.show()
```

```python
#OBJECT ORIENTED PLOT
#fig : entire figure, ax : individual plot
fig , ax = plt.subplots()
ax.plot(x,y)

ax.set_title("OBJECT ORIENTED PLOT")
ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')

plt.show()
```

```python
x = [1,2,3,4]
y1 = [10,20,25,30]
y2 = [30,25,20,10]

fig , axs = plt.subplots(1,2) # 1 row, 2 cloumn

axs[0].plot(x,y1, color='blue')
axs[0].set_title('First Plot')

axs[1].plot(x,y2, color='red')
axs[1].set_title('Second Plot')

plt.tight_layout()
plt.show()
```

```python
x = [1,2,3,4]
y1 = [10,20,25,30]
y2 = [30,25,20,10]

fig , axs = plt.subplots(2,1) # 1 row, 2 cloumn

axs[0].plot(x,y1, color='blue')
axs[0].set_title('First Plot')

axs[1].plot(x,y2, color='red')
axs[1].set_title('Second Plot')

plt.tight_layout()
plt.show()
```

```python
# Custom size & DPI - dot per inch
fig = plt.figure(figsize=(6,4), dpi=100)
plt.plot(x,y)
plt.title('Custom Size & DPI')
plt.show()
```

```python
#Save plot locally
plt.plot(x,y)
plt.title('Save this plot to local')
plt.savefig('plot.png', dpi=300)
plt.close()
```

```python
plt.plot(x,y)
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Time vs Value')
plt.legend() #Show the Label in plot
plt.show()
```

```python
plt.plot(x,y, label = 'Growth' )
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Time vs Value')
plt.legend() #Show the Label in plot
plt.show()
```

```python
plt.plot(x,y, color='green', linewidth = 2, linestyle='--')
plt.show()
```

```python
plt.plot(x,y, color='#a8328f', linewidth = 2, linestyle='--', marker='o', markersize='15' )
plt.show()
```

```python
plt.plot(x,y)
plt.xlim(0,10)
plt.ylim(0,50)
plt.show()
```

```python
# Bar graph
plt.bar(['A','B','C'], [5,3,9])
plt.title('Bar chart')
plt.show()
```

```python
plt.scatter(x,y, color = 'red')
plt.title('Scatter Plot')
plt.show()
```

```python
# Histogram
data = [1,2,3,3,4,5,5,5,6,6,7,8,8]
plt.hist(data, bins=8)
plt.title('Histogram')
plt.show()
```

```python
# PIE Chart
sizes = [30,25,45]
labels = ['A','B','C']
plt.pie(sizes, labels=labels, autopct = '%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
# PIE chart
sizes = [30, 45, 25]
labels = ['A', 'B', 'C']
colors = ['#ff9999','#66b3ff','#99ff99']
plt.pie(sizes, labels=labels, autopct='%1.0f%%', colors=colors)
plt.title('Pie Chart')
plt.show()
```
