```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('OnlineRetail.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
cust_data = data[data['CustomerID'].notna()]
cust_data
```

```python
cust_data.info()
```

```python
cust_data['InvoiceDate'] = cust_data['InvoiceDate'].astype(str) # use to convert in string
```

```python
cust_data['is_invoice_number_numeric'] = cust_data['InvoiceDate'].str.isdigit()
```

```python
non_numeric_invoice = cust_data[cust_data['is_invoice_number_numeric']==False]
```

```python
non_numeric_invoice
```

```python
# where ever the invoice no starts with C - it is cancelled orders
```

```python
cust_data = cust_data[cust_data['InvoiceNo'].str.startswith('C')==False]
```

```python
cust_data
```

```python
cust_data.info()
```

```python
cust_data['Total Transaction Amount'] = cust_data['Quantity'] * cust_data['UnitPrice']
```

```python
cust_data.head()
```

```python
# reference date - this is the date against which will shall measure the recency
# recency - how many days ago did the customer last make a purchase / transaction
```

```python
cust_data['InvoiceDate'] = pd.to_datetime(cust_data['InvoiceDate'])
```

```python
cust_data.info()
```

```python
ref_date = cust_data['InvoiceDate'].max() + pd.Timedelta(days = 1) # time delta is use to add X days to the date
```

```python
ref_date
```

```python
cust_data['CustomerID'] = cust_data['CustomerID'].astype(int)
```

```python
rfm = cust_data.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (ref_date - x.max()).days, #recency
    'InvoiceNo': 'nunique', # frequency
    'Total Transaction Amount': 'sum' # monetary
}).reset_index()
```

```python
rfm
```

```python
rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']
```
