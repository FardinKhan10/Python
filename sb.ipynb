```python
'''
Seaborn is a Python data visualization library based on matplotlib.
'''
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
tips.head()
tips.head(10)
```

```python
'''
Other dataset present
1. Tips Dataset
2. Iris Dataset
3. Penguins Dataset
4. Flights Dataset
5. Diamonds Dataset
6. Titanic Dataset
7. Exercise Dataset
8. MPG Dataset
9. Planets Dataset
'''
```

```python
# Histplot

sns.histplot(data = tips, x='total_bill', bins = 20, kde=True)
plt.title("Histogram of total bill with KDE")
plt.show()
```

```python
# KDEplot

sns.kdeplot(data=tips['tip'], fill=True, bw_adjust=0.5)
plt.title('KDE plot of tip amount')
plt.show()
```

```python
# new data
iris = sns.load_dataset('iris')
iris.head()
```

```python
# Displot
sns.displot(
    data=iris,
    x='sepal_length',
    hue = 'species',
    kind = 'kde',
    fill = True
)

plt.title('Sepal length distribution by species')
plt.show()
```

```python
# Barplot

sns.barplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='sex',
    ci='sd'
)
plt.title('Average Total Bill by Day and Gender')
plt.show()
```

```python
# Boxplot

sns.boxplot(
    data=tips,
    x='day',
    y='tip',
    hue='sex'    
)
plt.title('Boxplot of tip by Day and Gender')
plt.show()
```

```python
# Violinplot
sns.violinplot(
    data = tips,
    x='day',
    y='total_bill',
    hue= 'sex',
    split=True
)
plt.title('Violin plot of total bill')
plt.show()
```

```python
# Stripplot
sns.stripplot(
    data = tips,
    x='day',
    y='total_bill',
    hue='sex',
    jitter=True,
    dodge=True
)
plt.title('Raw tip data by day and gender')
plt.show()
```

```python
# Swarmplot

sns.swarmplot(
    data = tips,
    x='day',
    y='total_bill',
    hue='sex',
    dodge=True
)
plt.title('Swarmplot of bills')
plt.show()
```

```python
# Regplot

sns.regplot(
    data=tips,
    x='total_bill',
    y='tip'
)
plt.title('Total bill vs Tip with regression line')
plt.show()
```

```python
# LMplot

sns.lmplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='sex',
    col='smoker'
)
plt.title('LM plot, Gender & smoker wise Trend',y=1.05 )
plt.show()
```

```python
# PAIRplot

sns.pairplot(
    data=iris,
    hue='species'
)
plt.show()
```
